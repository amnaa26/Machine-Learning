```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score
```

```python
data = load_breast_cancer()
X = data.data
y = data.target

feature_names = data.feature_names
df = pd.DataFrame(X, columns=feature_names)
df.head()
```

```python
def cal_f_score(X,y):
    _, nfeatures = X.shape
    positive_indices = np.where(y==1)[0]
    negative_indices = np.where(y==0)[0]
    f_scores = []
    for i in range(nfeatures):
        total_mean = np.mean(X[:,i])
        positive_mean = np.mean(X[positive_indices, i])
        negative_mean = np.mean(X[negative_indices, i])
        numerator = (positive_mean - total_mean) ** 2 +(negative_mean - total_mean) ** 2
        var_positive = np.var(X[positive_indices,i])
        var_negative = np.var(X[negative_indices,i])
        denominator = var_positive + var_negative

        if denominator == 0:
            f_score = 0
        else:
            f_score = numerator / denominator    
        
        f_scores.append(f_score)
    return np.array(f_scores)

fscores = cal_f_score(X, y)
fscores
```
