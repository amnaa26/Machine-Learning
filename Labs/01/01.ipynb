## Task 2:

```python
#load the dataset
df = pd.read_csv('cancer.csv')

#c (unique categories in target variable)
print('Unique categories in target varaible [Level]: ', df['Level'].unique())
```

```python
#d
print("Total number of columns of entire dataset:", df.shape[1])
print("First 50 records:\n", df.head(50))
print("\n\nBottom 30 records:\n", df.tail(30))

print("Information related to the data:\n")
df.info()
print("\n\nDimension of the dataset: ", df.shape)
print("Size of the dataset: ", df.size)
```

```python
# e

print("Number of columns: ", len(df.columns))
print("Total number so sample: ", len(df))
```

```python
# f
#numerical
print("Number of records of each unique category of target varaiable", df['Level'].value_counts())

#graphical
sns.countplot(x="Level", data=df, hue = 'Level', palette="Set2")
plt.title("\nCount of Records per Category of Target Variable")
plt.xlabel("Level")
plt.ylabel("Count")
plt.show()
```

```python
# g

sub_sample = df.loc[550:900, 'Gender':'Coughing of Blood']
sub_sample.to_csv('sub_sample.csv', index=False)
print('Mode and Median:\n', sub_sample.median(), sub_sample.mean())
print("\n\n")
sub_sample.describe()
```

```python
# h

sub_sample2 = df.iloc[-20:, df.columns.get_loc('Shortness of Breath'):df.columns.get_loc('Frequent Cold')]
sub_sample2.to_csv('sub_sample2.csv', index=False)
sub_sample2.describe()
```

```python
# i

sample_record = df.iloc[[110,220,360,440,656,778,202,889]]
print("Average alcohol consumption:", sample_record['Alcohol use'].mean(), "\n\n")
sample_record.describe()
```
